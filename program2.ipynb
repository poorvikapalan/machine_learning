```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=sns.load_dataset("titanic")
print(data)
```

```python
print(data.head())
```

```python
x=data['age']
y=data['survived']
plt.figure()
plt.scatter(x,y)
plt.xlabel('age')
plt.ylabel('survived')
plt.title("scatter plot between the age and survived")
plt.show()
```

```python
x=data['age']
y=data['fare']
plt.figure()
plt.scatter(x,y)
plt.xlabel('age')
plt.ylabel('fare')
plt.title("scatter plot between the age and fare")
plt.show()
```

```python
pearson_correlation=x.corr(y)
print("person correlation cofficient",pearson_correlation)
```

```python
spearman_corr = x.corr(y, method='spearman')
print("\nSpearman Correlation:\n",spearman_corr)
```

```python
cov_matrix=data.cov(numeric_only=True)
print("covariance matrix:\n",cov_matrix)
```

```python
corr_matrix=data.corr(numeric_only=True)
print("correlation matrix:\n",corr_matrix)
```

```python
spearman_corr = data.corr(method='spearman',numeric_only=True)
print("\nSpearman Correlation matrix:\n",spearman_corr)
```

```python
plt.figure()
sns.heatmap(corr_matrix,annot=True)
plt.title("correlation matrix heatmap")
plt.show()
```
